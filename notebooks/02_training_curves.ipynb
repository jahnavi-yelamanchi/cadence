# 02 — Training Curves

Load W&B run history and plot:
- Train vs val loss
- Val FIR over epochs
- Learning rate schedule
- Val accuracy

```python
import wandb
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

sns.set_theme(style='darkgrid')
FIGURES = Path('../notebooks/figures')
FIGURES.mkdir(exist_ok=True)
```

```python
# Load the most recent cadence run from W&B
api = wandb.Api()
runs = api.runs('cadence')
run = runs[0]
history = run.history()
print(f"Run: {run.name}  |  {len(history)} steps")
history.head()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Loss
axes[0].plot(history['train_loss'], label='Train', color='#42a5f5')
axes[0].plot(history['val_loss'], label='Val', color='#ef5350')
axes[0].set_title('Loss')
axes[0].set_xlabel('Epoch')
axes[0].legend()

# FIR
axes[1].plot(history['val_FIR'], color='#ffa726', marker='o', markersize=4)
axes[1].set_title('Val False Interruption Rate (↓ better)')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('FIR')

# Accuracy
axes[2].plot(history['val_acc'], color='#66bb6a', marker='s', markersize=4)
axes[2].set_title('Val Accuracy (↑ better)')
axes[2].set_xlabel('Epoch')
axes[2].set_ylim(0, 1)

fig.suptitle('Cadence — Training Curves', fontsize=13)
fig.tight_layout()
fig.savefig(FIGURES / 'training_curves.png', dpi=150, bbox_inches='tight')
plt.show()
print(f"Best val FIR: {history['val_FIR'].min():.4f} at epoch {history['val_FIR'].idxmin() + 1}")
```
